### Data Set Description

This provides a quick description of the data set. This only goes over the training set and will cover its size, memory footprint, data type, summary statistics, and any distribution plots.

```r
# Open data set
madelonURL <- "http://archive.ics.uci.edu/ml/machine-learning-databases/madelon/MADELON/madelon_train.data"
data <- read.csv(url(madelonURL), sep = " ", header = FALSE)
data$V501 <- NULL #Gets ride of the final NA column
head(data)
```

```r
# Size of data set
dim(data)
```

```r
# Memory footprint
object.size(data)
```

```r
# Data Type
sapply(data, typeof)
```

```r
# Summary Statistics
dataMean <- apply(data, 2, mean)
dataMedian <- apply(data, 2, median)
dataSD <- apply(data, 2, sd)
```

```r
# Plots
hist(dataMean, col = 'light blue')
hist(dataMedian, col = 'light green')
plot(density(data.matrix(data)))
```
